```python
import numpy as np
```

```python
def convert(x):
    if 'M'in x:
        return float(x.replace("M",""))*1024
    elif 'k' in x:
     return float(x.replace("k",""))
    else:
     return np.nan
```

```python
df["Size"] = df["Size"].apply(convert)
```
