```python
import seaborn as sns
```

```python
df=sns.load_dataset("planets")
df.head(6)
```

```python
df.isna().sum()
```

```python
df.shape
```

```python
df.describe
```

```python
df.dtypes
```

```python
df["orbital_period"].unique
```

```python
df["orbital_period"]= df["orbital_period"].fillna(df["orbital_period"].mean())
df["orbital_period"]
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df["mass"].unique()
```

```python
df["mass"]=df["mass"].fillna(df["mass"].mean())
df["mass"]
df.isna().sum()
```

```python
df["distance"] = df["distance"].fillna(df["distance"].mean())
df["distance"]
df.isna().sum()
```

```python
import matplotlib.pyplot as plt
```

```python
sns.lineplot(x=df.index,y=df["number"],data=df)
plt.show()
```

```python
sns.lineplot(x=df.index,y=df["orbital_period"],data=df)
plt.show()
```

```python
sns.lineplot(x=df.index,y=df["distance"],data=df)
plt.show()
```

```python
sns.scatterplot(x=df["year"],y=df["distance"],data=df,hue='year',color='green')
plt.show()
```

```python
sns.scatterplot(x=df["number"],y=df["mass"],data=df,hue='year')
plt.show()
```

```python
sns.barplot(x=df.index,y=df["method"],data=df)
plt.show()
```
