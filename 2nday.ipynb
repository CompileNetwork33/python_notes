## slicing of list

```python
fruits = ["mango","apple","banana","grapes","orange"]
#start[0],step[1]by default
print(fruits[::2])
print(fruits[::-1])
```

```python
print(fruits[3::1])
```

```python
print(fruits[3:2:1])
print(fruits[3:2:-1])
```
