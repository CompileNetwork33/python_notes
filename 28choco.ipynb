```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\kashi\OneDrive\图片\Desktop\techcadd\chocolate_sales_2025_dataset.csv")
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['Brand'] = le.fit_transform(df['Brand'])
df['Product_Type']=le.fit_transform(df['Product_Type'])
df['Country'] = le.fit_transform(df['Country'])
df['Sales_Channel'] = le.fit_transform(df['Sales_Channel'])
df['Payment_Method'] = le.fit_transform(df['Payment_Method'])
df['Price_USD'] = le.fit_transform(df['Price_USD'])
df['Date'] = le.fit_transform(df['Date'])
df['Revenue_USD'] = le.fit_transform(df['Revenue_USD'])

```

```python
x = df[['Sale_ID', 'Date', 'Brand', 'Product_Type', 'Country', 'Sales_Channel',
       'Price_USD', 'Revenue_USD','Units_Sold']]
x.head()
       
    
```

```python
y=df['Payment_Method']
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
lr.fit(x_train,y_train)
```
